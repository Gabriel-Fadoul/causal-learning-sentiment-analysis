```python
import pandas as pd

# Exemplo: coluna contendo os sentimentos
# df = pd.read_csv("dados.csv")

def normalize_sentiment(sentiment):
    s = sentiment.strip().lower()

    if s in positive:
        return "positive"
    elif s in negative:
        return "negative"
    elif s in neutral:
        return "neutral"
    else:
        return "neutral"  # fallback padrão

# Aplicação
df_raw["sentiment_3class"] = df_raw["Sentiment"].apply(normalize_sentiment)

print(df_raw[["Sentiment", "sentiment_3class"]].head())
```

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained(
    "vinai/bertweet-base"
)

model = AutoModelForSequenceClassification.from_pretrained(
    "vinai/bertweet-base",
    num_labels=3
)
```

```python
label_map = {
    "negative": 0,
    "neutral": 1,
    "positive": 2
}

df_raw["label"] = df_raw["sentiment_3class"].map(label_map)
```

```python
from sklearn.model_selection import train_test_split

train_texts, test_texts, train_labels, test_labels = train_test_split(
    df_raw["Text"].tolist(),
    df_raw["label"].tolist(),
    test_size=0.2,
    random_state=42
)
```

```python
train_encodings = tokenizer(
    train_texts,
    truncation=True,
    padding=True,
    max_length=128
)

test_encodings = tokenizer(
    test_texts,
    truncation=True,
    padding=True,
    max_length=128
)
```

```python
import torch

class SentimentDataset(torch.utils.data.Dataset):

    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):

        item = {
            key: torch.tensor(val[idx])
            for key, val in self.encodings.items()
        }

        item["labels"] = torch.tensor(self.labels[idx])

        return item

    def __len__(self):
        return len(self.labels)
```

```python
train_dataset = SentimentDataset(
    train_encodings,
    train_labels
)

test_dataset = SentimentDataset(
    test_encodings,
    test_labels
)
```

```python
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

def compute_metrics(eval_pred):

    logits, labels = eval_pred

    predictions = np.argmax(logits, axis=-1)

    return {
        "accuracy": accuracy_score(labels, predictions),
        "f1": f1_score(labels, predictions, average="weighted")
    }
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs"
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics
)
```
